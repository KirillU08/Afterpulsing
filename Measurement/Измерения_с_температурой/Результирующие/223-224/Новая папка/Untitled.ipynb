```python
import pandas as pd
import io
import math
import matplotlib.pyplot as plt
import numpy as np
import string
from IPython.display import display
from numpy import random, mean, var, std
import os
```

```python
datasets=[]


#print(os.getcwd())
for d, dirs, files in os.walk(str(os.getcwd())+'\mes'):
    for f in files:
        ds=(np.loadtxt(os.getcwd()+'\mes\\'+f, dtype = str))
        datasets.append(ds)
        
DS_COUNT=len(datasets)
```

```python
pd_datasets=[]
for i in range(DS_COUNT):
    pd_dataset = pd.DataFrame(datasets[i][0:,0:], columns=['P','T(ns)','T(точки)','n','N','Exposition(ms)','Ch1','Ch2', 'SS', 'Temp'])
    display(pd_dataset)
    pd_datasets.append(pd_dataset)
```

```python
for i in range(DS_COUNT):
    ds=pd_datasets[i]
    DS_LEN=len(ds)
    #print(ds)
    for j in range(0, DS_LEN, 1):
        #pd_datasets[i][j][0]
        #print(ds.loc[j][0], ds.loc[j][1])
        NSS1=float(ds.loc[j][3])
        NK1=float(ds.loc[j][7])
        NSS2=float(ds.loc[j][8])
        NSS3=float(ds.loc[j][4])
        P=(NSS1*NK1-NSS2*NSS3)/(NSS3*(NK1-NSS3))
        #print(P)
        ds.set_value(j, 'P', str(P))
        #if(i==0):
            #display(ds)
```

```python
for i in range(DS_COUNT):
    print(i)
    with(open(str(DS_COUNT)+".csv", 'a')) as f:
        a=pd_datasets[i].to_csv(f, index=False)
        f.close()
```
