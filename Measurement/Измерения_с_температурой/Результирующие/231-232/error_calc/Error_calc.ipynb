```python
import pandas as pd
import io
import math
import matplotlib.pyplot as plt
import numpy as np
import string
from IPython.display import display
from numpy import random, mean, var, std
import os
```

```python
datasets=[]


print(os.getcwd())
for d, dirs, files in os.walk(str(os.getcwd())+'\mes'):
    for f in files:
        ds=(np.loadtxt(os.getcwd()+'\mes\\'+f, dtype = str))
        datasets.append(ds)
        
DS_COUNT=len(datasets)
```

```python
for i in range(DS_COUNT):
    data_pandas = pd.DataFrame(datasets[i][1:5,0:], columns=datasets[0][0, 0:])
    display(data_pandas)
```

```python
mean_arr=[]

COLUMN_NAMES=['Error','Time','Mean P']

for i in range(DS_COUNT):
    COLUMN_NAMES.append("P"+str(i+1))
    
DS_LEN=len(datasets[0])
resultDF = pd.DataFrame(columns = COLUMN_NAMES)


for i in range(1,DS_LEN,1):

    p_arr=[]
    time=float(datasets[0][i][1])
    for j in range(0, DS_COUNT, 1):
        ds_l=datasets[j]
        p=float(ds_l[i][0])
        p_arr.append(p)

    mean_p=np.mean(p_arr)
    std_p=np.std(p_arr)
    error=std_p/np.sqrt(DS_COUNT-1)

    data_pandas_res = pd.DataFrame(data=[{'Error': str(error), 'Time': time, 'Mean P': str(mean_p), 'P1': str(p_arr[0]), 'P2': str(p_arr[1]), 'P3': str(p_arr[2]), 'P4': str(p_arr[3])#, 'P5': str(p_arr[4]), 'P6': str(p_arr[5]), 'P7': str(p_arr[6]), 'P8': str(p_arr[7]), 'P9': str(p_arr[8]), 'P10': str(p_arr[9]), 'P11': str(p_arr[10]), 'P12': str(p_arr[11])
                                         }], columns = COLUMN_NAMES)
    resultDF = resultDF.append(data_pandas_res, ignore_index=True)
display(resultDF)
```

```python
f = open("result.csv", 'a')
a=resultDF.to_csv(f, index=False)
f.close()
```
